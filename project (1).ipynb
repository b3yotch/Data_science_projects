```python
df=pd.read_csv('/content/MagicBricks.csv')
df
```

```python
df1=df.drop(columns=["Status","Transaction","Per_Sqft"	],axis=1)
df1.sample(5)
```

```python
df1.sort_values("Price",inplace=True)
df1
```

```python
df1["Price"]=df1["Price"]/100000
```

```python
df1["Price per sqft"]=df1["Price"]*100000/df1["Area"]
df1.head()
```

```python
df1.dropna(inplace=True)
df1.isnull().sum()
```

```python
df2 = df1.copy()
value_counts = df2["Locality"].value_counts()
df2["Locality"] = df2["Locality"].apply(lambda x: "others" if value_counts[x] < 2 else x)
df2.head()
```

```python
df2["area_per"]=df2['Area']/df2["BHK"]
df2.head()
```

```python
def remove_outliers(df):
    q1 = df2["Price per sqft"].quantile(0.05)
    q3 = df2["Price per sqft"].quantile(0.95)
    df3 = df2[(df2["Price per sqft"] >= q1) & (df2["Price per sqft"] <= q3)]
    return df3

df3 = remove_outliers(df2)
```

```python
def remove_outliers(df):
    q1 = df3["area_per"].quantile(0.05)
    q3 = df3["area_per"].quantile(0.95)
    df4 = df3[(df3["area_per"] >= q1) & (df3["area_per"] <= q3)]
    return df4

df4 = remove_outliers(df3)
```

```python
df4=df4[df4["Bathroom"]<(df4["BHK"]+2)]
df4.shape
df4.head()
```

```python
df4.shape
```

```python
def plot_scatter_chart(df,Locality):
    bhk2 = df4[(df4.Locality==Locality) & (df4.BHK==2)]
    bhk3 = df4[(df4.Locality==Locality) & (df4.BHK==3)]
    plt.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.Area,bhk2.Price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.Area,bhk3.Price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(Locality)
    plt.legend()

plot_scatter_chart(df4,"others")
```

```python
def remove_out(df):
  max_bhk = df4['BHK'].max()
  correct = df4.copy()  # Create a copy of the original DataFrame

  for i in range(1, max_bhk):
        lower_bhk_mean = df4.loc[df['BHK'] == i, 'Price per sqft'].mean()
        higher_bhk_mask = (df4['BHK'] == i + 1) & (df4['Price per sqft'] < lower_bhk_mean)
        correct = correct.loc[~higher_bhk_mask]

  return correct

# Example usage
df5 = remove_out(df4)
```

```python
df5.shape
```

```python
def plot_scatter_chart(df,Locality):
    bhk2 = df5[(df5.Locality==Locality) & (df5.BHK==2)]
    bhk3 = df5[(df5.Locality==Locality) & (df5.BHK==3)]
    plt.rcParams['figure.figsize'] = (15,10)
    plt.scatter(bhk2.Area,bhk2.Price,color='blue',label='2 BHK', s=50)
    plt.scatter(bhk3.Area,bhk3.Price,marker='+', color='green',label='3 BHK', s=50)
    plt.xlabel("Total Square Feet Area")
    plt.ylabel("Price (Lakh Indian Rupees)")
    plt.title(Locality)
    plt.legend()

plot_scatter_chart(df4,"others")
```

```python
df6=df5[df5["Bathroom"]<(df5["BHK"]+2)]

df6.head()
```

```python
df7=df6.drop(["Furnishing","Type","Price per sqft","area_per","Parking"],axis=1)
df7.head(3)
```

```python
dummies = pd.get_dummies(df7.Locality)
dummies=dummies.astype(int)
dummies.head()
```

```python
df8 = pd.concat([df7,dummies.drop('others',axis='columns')],axis='columns')
df8.head()
```

```python
df9 = df8.drop('Locality',axis='columns')
df9.head(2)
```

```python
x=df9.drop('Price',axis='columns')
y=df9.Price
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import Ridge
r_clf = Ridge()
r_clf.fit(x_train,y_train)
r_clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import Lasso, Ridge

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

cross_val_score(Ridge(), x, y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor


from sklearn.tree import DecisionTreeRegressor
def find_best_score(x,y):
  algos={
      "Lasso":{
          "model":Lasso(),
          "params":{
              "alpha":[1,2],
              "selection":['random','cyclic']
          }
      },
      "Decision Tree":{
          "model":DecisionTreeRegressor(),
          "params":{
              'criterion':['mse','friedman_mse'],
              'splitter':['best','random']
          }
      },
      "Ridge":{
          "model":Ridge(),
          "params":{
              'alpha':[1,2],
              'solver':['auto','svd']
          }
      },
      "Random Forest":{
          "model":RandomForestRegressor(),
          "params":{
              'n_estimators':[1,5,10]
          }
      }
  }
  scores=[]
  cv=ShuffleSplit(n_splits=5,test_size=0.2,random_state=0)
  for algos_name,config in algos.items():
    gs=GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False)
    gs.fit(x,y)
    scores.append({
        'model':algos_name,
        'best_score':gs.best_score_,
        'best_params':gs.best_params_
    })
  return pd.DataFrame(scores,columns=['model','best_score','best_params'])
find_best_score(x,y)



```

```python
def predict_price(Locality,Area,BHK,Bathroom):
    loc_index = np.where(x.columns==Locality)[0][0]

    x1 = np.zeros(len(x.columns))
    x1[0] = Area
    x1[1] = Bathroom
    x1[2] = BHK
    if loc_index >= 0:
        x1[loc_index] = 1

    return r_clf.predict([x1])[0]
```
